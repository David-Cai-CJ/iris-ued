```python
from iris.time_series import autocorr1d
import matplotlib.pyplot as plt
import numpy as n

%matplotlib inline
```

# Visualization of autocorrelation on typical signals

```python
x = n.linspace(0, 10, num = 512)
signal = n.ones_like(x)
signal[100:] = n.exp(-1*x[:-100]) + 0.01*n.sin(n.pi * x[:-100])

fig = plt.figure()
ax = fig.add_subplot(211)
ax.plot(x, signal)

ax2 = fig.add_subplot(212)
ax2.plot(autocorr1d(signal))
```
